```python
%matplotlib inline

files = ['test-1-EO', 'test-2-EC', 'test-3-EO', 'test-4-EC', 'test-5-nap']
pathlib = '/Users/ethan/Dropbox/Documents/Research/Projects_current/COBE-EEG-1/functions'
pathdata = '/Users/ethan/Dropbox/Documents/Research/Projects_current/COBE-EEG-1/data/test-2'
import pylab as plt




os.chdir(pathdata)
for file in files:
    df = pd.read_csv(file + '.csv', sep=',')
    data = df.as_matrix()
    
    
    
    chans = [11]
    low_cutoff = 5
    high_cutoff = 20
    
    for s in chans:
        
        # get time-stamp data
        t = data[:,0]
        y = data[:,s]
        Y    = np.fft.fft(y)
        freq = np.fft.fftfreq(len(y), t[1] - t[0])

        # remove DC (non-varying) component
        freq = freq[1:]
        Y = np.abs(Y[1:])

        pylab.figure()
        pylab.plot( freq, Y )
        #plt.xlabel('Channel '+str(s))
        plt.title(file)

        #pp.savefig()
        #yheight = max(freq[low_cutoff:high_cutoff])

        plt.xlim(low_cutoff,high_cutoff)
        #plt.ylim(0,0.2)




```
